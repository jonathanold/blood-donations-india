## Loop

Here, we loop over all state-district pairs. We send a request to the server and collect the data in a list. The final dataset is saved with today's name as date.

```python
print(state_district_list)
```

```python
base_url = "https://www.eraktkosh.mohfw.gov.in/BLDAHIMS/bloodbank/nearbyBB.cnt?hmode=GETNEARBYSTOCKDETAILS&stateCode={}&districtCode={}&bloodGroup=all&bloodComponent=11&lang=0&_=1707784201003"

for stateid, distid in tqdm(state_district_list, desc="Processing states and districts"):

    url = base_url.format(stateid, distid)
    response = requests.post(url, headers=headers, verify=False)
    response_data = json.loads(response.text)
    if 'data' in response_data and not response_data['data']:
        print(f"State={stateid}, distid={distid} does not exist.")
    else:
        url = base_url.format(stateid, distid)

        response = requests.post(url, headers=headers, verify=False)
        response_data = json.loads(response.text)    
        # Assuming response_data is the JSON response
        blood_banks = response_data['data']

        # Initialize an empty list to store individual blood bank data
        blood_bank_data_list = []

        # Iterate through each blood bank entry
        for blood_bank_info in blood_banks:
            # Extract information for each blood bank
            blood_bank_name = blood_bank_info[1].split("<br/>")[0]
            blood_bank_address = blood_bank_info[1].split("<br/>")[1]
            # Extract phone, fax, email, etc. as shown before
            contact_info = blood_bank_info[1].split('<br/>')[2]

            # Check if there is 'Phone:' in the contact_info
            if 'Phone: ' in contact_info:
                if 'Fax' in contact_info:
                    phone_number = contact_info.split('Phone: ')[1].split(' ,Fax')[0]
                else:
                    phone_number = contact_info.split('Phone: ')[1].split(' ,Email')[0]

            else:
                phone_number = None  # or handle the case when 'Phone:' is not found
            
            # Check if there is 'Email:' in the contact_info
            if 'Email: ' in contact_info:
                email = contact_info.split('Email: ')[1]
            else:
                email = None  # or handle the case when 'Email:' is not found

            # Extract the category of the blood bank
            category = blood_bank_info[2]

            # Extract the availability of blood stock
            availability_info = blood_bank_info[3]
            availability = availability_info.split('>')[1].split('<')[0]

            # Extract the last update of the stock
            last_update = blood_bank_info[4]

            # Extract the type of institution
            institution_type = blood_bank_info[5]

            # Append the extracted information to the list
            blood_bank_data_list.append({
                "State ID": stateid,
                "District ID": distid,
                "Name": blood_bank_name,
                "Address": blood_bank_address,
                "Phone": phone_number,
                "Email": email,
                "Category": category,
                "Availability": availability,
                "Last Update": last_update,
                "Institution Type": institution_type
                
            })

        # Assuming blood_bank_data_list is the list of dictionaries
        df_new = pd.DataFrame(blood_bank_data_list)
        df = pd.concat([df, df_new], ignore_index=False)

        tqdm.write(f"Got data for state={stateid}, distid={distid}.")

# Get the current date and time
current_date = datetime.now().strftime("%Y-%m-%d")
# Create a file name using the formatted date
file_name = f"../../Data/India Blood Banks/eRaktKosh_scraped/eRaktKosh_scraped_{current_date}.csv"

df.to_csv(file_name, index=False)
```
